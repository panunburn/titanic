```python
import numpy as np
import pandas as pd
from sklearn import linear_model
import matplotlib as plt
```

```python
name = pd.read_csv('train.csv')
```

```python
select = pd.DataFrame(name,columns = ['Pclass','Sex','Age','Parch','Fare','SibSp'])
```

```python
select
```

```python
select['Sex'] = np.where(select['Sex'] == 'male', 0, 1)
```

```python
select['Sex']
```

```python
select['Age'].fillna(select['Age'].mean(),inplace = True)
```

```python
select.describe()
```

```python
live = name['Survived']
regre = linear_model.LinearRegression()
```

```python
regre.fit(select, live)
regre.score(select, live)
```
